* Reverse word order from the input string 


```python
# with list append
teststring='this is first time'
result = teststring.split(" ")
rev=[]

for i in result[::-1]:
    rev.append(i)

' '.join(rev)
```

```python
#with list insert
teststring='this is first time'
def reverse_v1(x):
    y = x.split()
    result = []
    for word in y:
        result.insert(0,word)
    return " ".join(result)

reverse_v1(teststring)
```

```python
# with list reverse elements

teststring='this is first time'

def reverse_v2(x):
  y = x.split()
  return " ".join(y[::-1])

reverse_v2(teststring)
```

```python
#with reversed function over list
def reverse_v3(x):
    y = x.split()
    return " ".join(reversed(y))

reverse_v3(teststring)
```

```python
#with reverse method over list

def reverse_v4(x):
  y = x.split()
  y.reverse()
  return " ".join(y)

reverse_v4(teststring)
```

* Reverse word order from the input string by using for comprehension 


```python
teststring='this is first time'
result = teststring.split(" ")
rev=[]
for i in result[::-1]:
       rev.append(i)
print (" ".join(rev))
```

```python
teststring='this is first time'
result = teststring.split(" ")
rev = [ item for item in result[::-1] ]
print (" ".join(rev))
```

```python
teststring='this is first time'
print (" ".join([ item for item in teststring.split(" ")[::-1] ]))
```
