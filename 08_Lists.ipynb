__Anonimous or lambda functions__  
simple functions consisting of a single statement, the result of which is the return value.  
defined using the lambda keyword, which has no meaning other than “we are declaring an anonymous function.”  



```python
def short_funct(x):
    return x*2
```

```python
the_same_short_function=lambda x: x*2
```

```python
short_funct(3)
```

```python
the_same_short_function(3)
```

# List comprehensions

* allow to concisely form a new list by filtering the elements of a collection and transforming the elements passing the filter in one concise expression.


```python
a=[1,2,3,4, 5, 6]
result = []
for val in a:
    if val%2:
        result.append(val**2)
        
result
```

```python
result2=[ val**2 for val in a if val%2      ]
```

```python
result2
```

```python
result2=[ val**2 if val%2 else 1 for val in a  ]
```

```python
result2
```

```python
result2=[ val**2 if val%2 for val in a  ]
```

```python
strings=['a','dffd','vb','fdsfdsdfgfds']
[ x.upper() for x in strings if len(x)>2 ]
```

# Nested list Comprehension

* nested list comprehensions are a bit hard to wrap your head around. 
* The for parts of the list comprehension are arranged according to the order of nesting, 
* filter condition is put at the end as before. 
* example
  * we “flatten” a list of tuples of integers into a simple list of integers:



```python
some_tuples=[(1,2,3),(4,5,6),(7,8,9)]
flatten=[]
for tup in some_tuples:
    for element in tup:
        flatten.append(element)
flatten
```

```python
flatten2=[element for tup in some_tuples for element in tup ]
flatten2
```

  * Keep in mind that the order of the for expressions would be the same if you wrote a nested for loop instead of a list comprehension
